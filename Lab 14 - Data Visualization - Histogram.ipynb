## Hands-on Lab: Visualizing Data with Histograms


### 1. Visualizing the distribution of data (Histograms)


**1.1 Histogram of `CompTotal` (Total Compensation)**


Objective: Plot a histogram of `CompTotal` to visualize the distribution of respondents' total compensation.


```python
## Write your code here
QUERY = "SELECT CompTotal FROM main WHERE CompTotal IS NOT NULL AND CompTotal < 1000000;"
data = pd.read_sql_query(QUERY, conn)

plt.hist(data['CompTotal'], bins=50, color='red', edgecolor='black')
plt.title('Total Comp. - Distribution')
plt.xlabel('CompTotal')
plt.ylabel('Frequency')
plt.show()
```

**1.2 Histogram of YearsCodePro (Years of Professional Coding Experience)**


Objective: Plot a histogram of `YearsCodePro` to analyze the distribution of coding experience among respondents.


```python
!pip install seaborn
import seaborn as sns
```

```python
## Write your code here
query = "SELECT YearsCodePro FROM main WHERE YearsCodePro IS NOT NULL;"
df = pd.read_sql_query(query, conn)
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

plt.figure(figsize=(10, 6))
sns.histplot(df['YearsCodePro'].dropna(), bins=20, kde=True, color='red')
plt.title('YearsCodePro - Histogram Graph')
plt.xlabel('Years of CodePro')
plt.ylabel('Frequency')
plt.show()
```

### 2. Visualizing Relationships in Data


**2.1 Histogram Comparison of `CompTotal` by `Age` Group**


Objective: Use histograms to compare the distribution of CompTotal across different Age groups.


```python
## Write your code here
query = """
SELECT 
    CompTotal, 
    Age,
    CASE 
        WHEN Age = 'Under 18 years old' THEN 15
        WHEN Age = '18-24 years old' THEN 21
        WHEN Age = '25-34 years old' THEN 30
        WHEN Age = '35-44 years old' THEN 40
        WHEN Age = '45-54 years old' THEN 50
        WHEN Age = '55-64 years old' THEN 60
        WHEN Age = '65 years or older' THEN 70
        ELSE NULL
    END AS AgeNumeric
FROM main
WHERE CompTotal IS NOT NULL AND Age IS NOT NULL;
"""
df = pd.read_sql_query(query, conn)
df['CompTotal'] = pd.to_numeric(df['CompTotal'], errors='coerce')
df = df[df['CompTotal'] <= 1_000_000]

plt.figure(figsize=(15, 8))
sns.histplot(
    data=df, 
    x='CompTotal', 
    hue='Age',
    bins=30, 
    kde=False, 
    element='step'
)
plt.title('Comparison of CompTotal by Age')
plt.xlabel('Total Comp.')
plt.ylabel('Frequency')
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**2.2 Histogram of TimeSearching for Different Age Groups**


Objective: Use histograms to explore the distribution of `TimeSearching` (time spent searching for information) for respondents across different age groups.


```python
## Write your code here
query = "SELECT Age, TimeSearching FROM main WHERE TimeSearching IS NOT NULL AND Age IS NOT NULL;"
df = pd.read_sql_query(query, conn)
plt.figure(figsize=(15, 8))
sns.histplot(data=df, x='TimeSearching', hue='Age', bins=20, kde=False, element='step')
plt.title('TimeSearching for AgeGroups - Histogram Graph')
plt.xlabel('TimeSearching')
plt.ylabel('Frequency')
plt.legend(title='Age Group')
plt.show()
```

### 3. Visualizing the Composition of Data


**3.1 Histogram of Most Desired Databases (`DatabaseWantToWorkWith`)**


Objective: Visualize the most desired databases for future learning using a histogram of the top 5 databases.


```python
## Write your code here
query = """
SELECT DatabaseWantToWorkWith, COUNT(*) AS Count 
FROM main 
GROUP BY DatabaseWantToWorkWith 
ORDER BY Count DESC 
LIMIT 5;
"""
df = pd.read_sql_query(query, conn)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, y='DatabaseWantToWorkWith', x='Count', palette='coolwarm')
plt.title('Database Want To Work With')
plt.xlabel('Type of Databases')
plt.ylabel('No. of Respondants')
plt.show()
```

**3.2 Histogram of Preferred Work Locations (`RemoteWork`)**


Objective: Use a histogram to explore the distribution of preferred work arrangements (`remote work`).


```python
## Write your code here
query = "SELECT RemoteWork, COUNT(*) AS Count FROM main GROUP BY RemoteWork ORDER BY Count DESC;"
df = pd.read_sql_query(query, conn)

plt.figure(figsize=(15, 8))
sns.barplot(data=df, x='Count', y='RemoteWork', palette='YlGnBu')
plt.title('Preferred Work Locations')
plt.xlabel('No. of Respondants')
plt.ylabel('Category of RemoteWork')
plt.show()
```

### 4. Visualizing Comparison of Data


**4.1 Histogram of Median CompTotal for Ages 45 to 60**


Objective: Plot the histogram for `CompTotal` within the age group 45 to 60 to analyze compensation distribution among mid-career respondents.


```python
## Write your code here
query = """
SELECT Age, CompTotal 
FROM main 
WHERE CompTotal IS NOT NULL 
  AND Age IN ('45-54 years old', '55-64 years old');
"""
df = pd.read_sql_query(query, conn)
df['CompTotal'] = pd.to_numeric(df['CompTotal'], errors='coerce')

plt.figure(figsize=(15, 8))
sns.histplot(data=df, x='CompTotal', bins=20, kde=True, color='red')
plt.title('Median CompTotal for Ages 45-60 - Histogram Graph')
plt.xlabel('Total Comp.')
plt.ylabel('Frequency')
plt.show()
```

**4.2 Histogram of Job Satisfaction (`JobSat`) by YearsCodePro**


Objective: Plot the histogram for `JobSat` scores based on respondents' years of professional coding experience.


```python
## Write your code here
query = "SELECT YearsCodePro, JobSat FROM main WHERE JobSat IS NOT NULL AND YearsCodePro IS NOT NULL;"
df = pd.read_sql_query(query, conn)
plt.figure(figsize=(15, 8))
sns.histplot(data=df, x='JobSat', hue='YearsCodePro', bins=20, kde=False, element='step', palette='magma')
plt.title('JobSat by YearsCodePro - Histogram Graph')
plt.xlabel('JobSat')
plt.ylabel('Frequency')
plt.legend(title='Years of Experience', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
